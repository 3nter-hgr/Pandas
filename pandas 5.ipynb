```python
import pandas as pd
```

```python
movies =pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[(movies.duration >200) & (movies.genre == 'Crime')] #& <-- and
```

```python
movies[(movies.duration >200) | (movies.genre == 'Crime')] #| <-- or
```

```python
(movies.duration >200) | (movies.genre == 'Crime') #<-- will return boolean values
```

```python
(movies.duration >200) & (movies.genre == 'Crime')
```

```python
movies[(movies.genre=='Crime') | (movies.genre =='Drama') | (movies.genre =='Action')]
```

```python
movies.genre.isin(['Crime' , 'Action' , 'Drama'])
```

```python
movies[movies['genre'].isin(['Crime' , 'Action' , 'Drama'])]
```
