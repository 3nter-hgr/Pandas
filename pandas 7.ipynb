```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head(10)
```

```python
booleans =[]
for length in movies.duration:
    if length >=200:
        booleans.append(True)
    else:
        booleans.append(False)
booleans[0:10]
```

```python
movies.shape
```

```python
len(booleans)
# it makes sense 
#number of rows equals the number of booleans in the list
```

```python
#we will convert booleans list to series
is_long =pd.Series(booleans)
is_long
```

```python
movies[is_long]
```

```python
#another shorter way
is_long = movies.duration >= 200
movies[is_long]
```

```python
#another way
movies[movies.duration >= 200]
```

```python
movies[movies.duration >= 200].genre
#we want to know the genre of movies which its duration >= 200
```

```python
#another way to do that
movies[movies.duration >= 200]['genre']
```

```python
#loc allows you to select rows and columns by labels
movies .loc[movies.duration >= 200 , 'genre']
            #rows                    #column
```
