```python
pd.read_table('http://bit.ly/movieusers')
#as you can see the dataframe is tab seperated
```

```python
#you can avoid that by : 
pd.read_table('http://bit.ly/movieusers' , sep ='|')
#yay ot worked!
```

```python
# another problem is that the first row become the header and that is NOT true
# so we need to tell pandas that there is no header!
pd.read_table('http://bit.ly/movieusers' ,sep = '|' ,header =None)
```

```python
#okey we now want to name the header
user_cols = ['id' , 'age' , 'gender' , 'occupation' , 'zip_code']
pd.read_table('http://bit.ly/movieusers' ,sep = '|' , names =user_cols)
#it worked!
```

```python
users=pd.read_table('http://bit.ly/movieusers' , sep ='|' , names = user_cols , skiprows=1)
#skips the first row
```

```python
#if you want the first five rows
users.head()
```

```python
#if you want the last five rows
users.tail()
```

```python
#if you want any number of rows
users.head(15)
```
